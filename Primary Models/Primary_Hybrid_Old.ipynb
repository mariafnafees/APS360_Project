Hybrid CNN (mobilenetv2) + RNN (GRU)

# Reduced these parameters in order to run faster
- batch_size=8  # Instead of 32
- image_size=128  # In strokes_to_image function
- num_epochs=5

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torch.nn.functional as F
import numpy as np
import os
import time
import matplotlib.pyplot as plt
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, models
import random
from sklearn.metrics import confusion_matrix
import seaborn as sns
from PIL import Image, ImageDraw
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# Function to convert stroke data to images
def strokes_to_image(stroke_data, image_size=96):
    """Convert stroke data to image for CNN processing"""
    img = Image.new('L', (image_size, image_size), 'white')
    draw = ImageDraw.Draw(img)

    if stroke_data is None or len(stroke_data) == 0:
        return np.array(img)

    # fit stroke coordinates to image size
    scale_factor = image_size - 10

    # extract all x and y coordinates
    all_x = []
    all_y = []

    for point in stroke_data:
        if len(point) >= 2:
            all_x.append(point[0])
            all_y.append(point[1])

    if not all_x or not all_y:
        return np.array(img)

    min_x, max_x = min(all_x), max(all_x)
    min_y, max_y = min(all_y), max(all_y)

    #  scaling factors
    x_range = max_x - min_x
    y_range = max_y - min_y

    if x_range == 0: x_range = 1
    if y_range == 0: y_range = 1

    scale_x = scale_factor / x_range
    scale_y = scale_factor / y_range
    scale = min(scale_x, scale_y)

    # calculate offset- center the drawing
    offset_x = (image_size - scale * x_range) / 2 - scale * min_x
    offset_y = (image_size - scale * y_range) / 2 - scale * min_y

    # Draw strokes
    current_stroke = []
    for point in stroke_data:
        if len(point) >= 3:  #
            x, y, pen = point[0], point[1], point[2]

            # Transform coordinates
            screen_x = int(x * scale + offset_x)
            screen_y = int(y * scale + offset_y)

            if pen == 1:
                current_stroke.append((screen_x, screen_y))
            elif len(current_stroke) > 0:
                if len(current_stroke) == 1:
                    draw.point(current_stroke[0], fill='black')
                else:  # Multiple points
                    draw.line(current_stroke, fill='black', width=2)
                current_stroke = [(screen_x, screen_y)]

    if len(current_stroke) > 0:
        if len(current_stroke) == 1:
            draw.point(current_stroke[0], fill='black')
        else:
            draw.line(current_stroke, fill='black', width=2)

    return np.array(img)

# simple RNN branch (GRU)
class RNNBranch(nn.Module):
    def __init__(self, input_size=3, hidden_size=128):
        super(RNNBranch, self).__init__()
        self.hidden_size = hidden_size
        self.gru = nn.GRU(input_size, hidden_size, batch_first=True)

    def forward(self, x, lengths=None):
        batch_size = x.size(0)
        h0 = torch.zeros(1, batch_size, self.hidden_size).to(x.device)

        # Forward pass
        _, h_n = self.gru(x, h0)

        # Return the final hidden state
        return h_n.squeeze(0)

# MobileNetV2 cnn + RNN hybrid model
class MobileNetRNNHybrid(nn.Module):
    def __init__(self, num_classes=9, use_pretrained=True):
        super(MobileNetRNNHybrid, self).__init__()
        self.name = "mobilenet_rnn_hybrid"

        # CNN branch
        self.mobilenet = models.mobilenet_v2(pretrained=use_pretrained)
        self.mobilenet.classifier = nn.Identity()

        # RNN branch + stroke data
        self.rnn_branch = RNNBranch(input_size=3, hidden_size=128)

        # fusion layer
        self.fusion = nn.Sequential(
            nn.Linear(1280 + 128, 256),  # 1280 from MobileNetV2, 128 from RNN
            nn.ReLU(),
            nn.Dropout(0.3)
        )

        # classifier
        self.classifier = nn.Linear(256, num_classes)

    def forward(self, image_input, stroke_input, stroke_lengths=None):
        # Extract features

        cnn_features = self.mobilenet(image_input)  # [batch_size, 1280]

        rnn_features = self.rnn_branch(stroke_input, stroke_lengths)  # [batch_size, 128]

        combined_features = torch.cat((cnn_features, rnn_features), dim=1)

        fused_features = self.fusion(combined_features)

        # Final classification
        logits = self.classifier(fused_features)

        return logits, fused_features

def pad_stroke_sequences(sequences):
    """Pad sequences to the same length for batch processing"""
    max_len = min(100, max(len(seq) for seq in sequences))

    point_dim = sequences[0].shape[1] if len(sequences[0].shape) > 1 else 3

    #  padd sequences
    padded_sequences = torch.zeros(len(sequences), max_len, point_dim)
    lengths = torch.zeros(len(sequences), dtype=torch.long)

    for i, seq in enumerate(sequences):
        if len(seq) > 0:
            seq_tensor = torch.FloatTensor(seq)
            seq_len = min(max_len, seq_tensor.size(0))
            padded_sequences[i, :seq_len] = seq_tensor[:seq_len]
            lengths[i] = seq_len
        else:
            lengths[i] = 0

    return padded_sequences, lengths

class QuickDrawDataset(Dataset):
    def __init__(self, stroke_data, labels, transform=None):
        self.stroke_data = stroke_data
        self.labels = labels
        self.transform = transform

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        # get stroke data
        stroke_seq = self.stroke_data[idx]
        label = self.labels[idx]

        # stroke data to image
        img = strokes_to_image(stroke_seq)

        if self.transform:
            img = self.transform(img)

        # Convert stroke sequence to tensor
        if stroke_seq is not None and len(stroke_seq) > 0:
            stroke_seq = np.array(stroke_seq, dtype=np.float32)
        else:
            stroke_seq = np.zeros((1, 3), dtype=np.float32)

        return img, stroke_seq, label

def collate_fn(batch):
    images, stroke_seqs, labels = zip(*batch)

    images = torch.stack(images)
    labels = torch.LongTensor(labels)

    # Pad stroke sequences
    padded_seqs, lengths = pad_stroke_sequences(stroke_seqs)

    return images, padded_seqs, lengths, labels

# training function
def train(model, train_dataset, val_dataset, device,
          batch_size=32, num_epochs=10, learning_rate=0.01):
    """
    Train the model following the same approach as the baseline model
    """
    train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True,
                              pin_memory=True, collate_fn=collate_fn)
    val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False,
                           pin_memory=True, collate_fn=collate_fn)

    # Loss function
    criterion = nn.CrossEntropyLoss()


    optimizer = optim.SGD(model.parameters(), lr=learning_rate, momentum=0.9)

    scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, mode='max', factor=0.5,
                                                    patience=2, verbose=True)

    #  tracking metrics
    train_losses = []
    train_accs = []
    val_losses = []
    val_accs = []
    best_val_acc = 0

    # training
    start_time = time.time()

    for epoch in range(num_epochs):
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        train_pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Train]")
        for imgs, stroke_seqs, lengths, labels in train_pbar:
            imgs = imgs.to(device)
            stroke_seqs = stroke_seqs.to(device)
            lengths = lengths.to(device)
            labels = labels.to(device)

            optimizer.zero_grad()

            # Forward pass
            outputs, _ = model(imgs, stroke_seqs, lengths)
            loss = criterion(outputs, labels)

            # Backward pass
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

            train_pbar.set_postfix({
                'loss': f"{running_loss / (train_pbar.n + 1):.4f}",
                'acc': f"{100. * correct / total:.2f}%"
            })

        # training metrics
        train_loss = running_loss / len(train_loader)
        train_acc = 100. * correct / total
        train_losses.append(train_loss)
        train_accs.append(train_acc)


        model.eval()
        running_loss = 0.0
        correct = 0
        total = 0

        with torch.no_grad():
            val_pbar = tqdm(val_loader, desc=f"Epoch {epoch+1}/{num_epochs} [Valid]")
            for imgs, stroke_seqs, lengths, labels in val_pbar:
                imgs = imgs.to(device)
                stroke_seqs = stroke_seqs.to(device)
                lengths = lengths.to(device)
                labels = labels.to(device)

                outputs, _ = model(imgs, stroke_seqs, lengths)
                loss = criterion(outputs, labels)

                running_loss += loss.item()
                _, predicted = outputs.max(1)
                total += labels.size(0)
                correct += predicted.eq(labels).sum().item()

                val_pbar.set_postfix({
                    'loss': f"{running_loss / (val_pbar.n + 1):.4f}",
                    'acc': f"{100. * correct / total:.2f}%"
                })

        #  validation metrics
        val_loss = running_loss / len(val_loader)
        val_acc = 100. * correct / total
        val_losses.append(val_loss)
        val_accs.append(val_acc)

        scheduler.step(val_acc)

        print(f"Epoch {epoch+1}/{num_epochs}: "
              f"Train Loss: {train_loss:.4f}, Train Acc: {train_acc:.2f}%, "
              f"Val Loss: {val_loss:.4f}, Val Acc: {val_acc:.2f}%")

        # save model if validation accuracy improves
        if val_acc > best_val_acc:
            best_val_acc = val_acc
            torch.save(model.state_dict(), f"{model.name}_best.pth")
            print(f"New best model saved with validation accuracy: {val_acc:.2f}%")

    #  total training time
    total_time = time.time() - start_time
    print(f"Training completed in {total_time:.2f} seconds")

    #  training curves
    epochs = list(range(1, num_epochs + 1))
    plt.figure(figsize=(12, 5))

    plt.subplot(1, 2, 1)
    plt.plot(epochs, train_losses, 'b-', label='Train Loss')
    plt.plot(epochs, val_losses, 'r-', label='Val Loss')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()

    plt.subplot(1, 2, 2)
    plt.plot(epochs, train_accs, 'b-', label='Train Acc')
    plt.plot(epochs, val_accs, 'r-', label='Val Acc')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy (%)')
    plt.legend()

    plt.tight_layout()
    plt.savefig(f"{model.name}_training_curves.png")
    plt.show()

    return model, train_losses, train_accs, val_losses, val_accs

# evaluate model accuracy
def evaluate(model, data_loader, device):
    model.eval()
    correct = 0
    total = 0

    with torch.no_grad():
        for imgs, stroke_seqs, lengths, labels in tqdm(data_loader, desc="Evaluating"):
            imgs = imgs.to(device)
            stroke_seqs = stroke_seqs.to(device)
            lengths = lengths.to(device)
            labels = labels.to(device)

            outputs, _ = model(imgs, stroke_seqs, lengths)
            _, predicted = outputs.max(1)

            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    accuracy = 100. * correct / total
    return accuracy

#  confusion matrix
def plot_confusion_matrix(model, test_loader, class_names, device):
    model.eval()
    all_predictions = []
    all_labels = []

    with torch.no_grad():
        for imgs, stroke_seqs, lengths, labels in tqdm(test_loader, desc="Generating predictions"):
            imgs = imgs.to(device)
            stroke_seqs = stroke_seqs.to(device)
            lengths = lengths.to(device)

            outputs, _ = model(imgs, stroke_seqs, lengths)
            _, predictions = outputs.max(1)

            all_predictions.extend(predictions.cpu().numpy())
            all_labels.extend(labels.numpy())

    cm = confusion_matrix(all_labels, all_predictions)

    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
    plt.xlabel('Predicted')
    plt.ylabel('True')
    plt.title('Confusion Matrix')
    plt.savefig(f"{model.name}_confusion_matrix.png")
    plt.show()

    return cm

# main function to run everything
def run_model(data_dir=None):
    if data_dir is None:
        if os.path.exists('/content/drive'):
            data_dir = '/content/drive/MyDrive/APS360 Project/*Data Processing/Stroke order samples'
        else:
            data_dir = '.'  # Current directory

    # check for GPU
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    print(f"Using device: {device}")

    # load QuickDraw data
    print("Loading data...")
    try:
        X_data = np.load(os.path.join(data_dir, "X_data_full.npy"), allow_pickle=True)
        y_data = np.load(os.path.join(data_dir, "y_data_full.npy"))
        print(f"Loaded data from {data_dir}")
    except FileNotFoundError:
        print(f"Could not find data in {data_dir}, trying current directory...")
        try:
            X_data = np.load("X_data_full.npy", allow_pickle=True)
            y_data = np.load("y_data_full.npy")
            print("Loaded data from current directory")
        except FileNotFoundError:
            raise FileNotFoundError("Could not find data")

    # load category mapping
    try:
        import json
        mapping_file = os.path.join(data_dir, "category_mapping.json")
        if os.path.exists(mapping_file):
            with open(mapping_file, 'r') as f:
                category_mapping = json.load(f)
            # Convert to list for indexing
            CLASS_NAMES = [k for k, v in sorted(category_mapping.items(), key=lambda item: item[1])]
        else:
            raise FileNotFoundError
    except:

        print("Category mlass names...")
        CLASS_NAMES = [f"class_{i}" for i in range(len(set(y_data)))]

    print(f"Classes: {CLASS_NAMES}")
    print(f"Total samples: {len(X_data)}")

    # split data into train, validation, and test sets
    X_train_val, X_test, y_train_val, y_test = train_test_split(
        X_data, y_data, test_size=0.15, random_state=42, stratify=y_data)

    X_train, X_val, y_train, y_val = train_test_split(
        X_train_val, y_train_val, test_size=0.17647, random_state=42, stratify=y_train_val)

    print(f"Train samples: {len(X_train)}")
    print(f"Validation samples: {len(X_val)}")
    print(f"Test samples: {len(X_test)}")

    # image transformations
    transform = transforms.Compose([
        transforms.ToPILImage(),
        transforms.Resize((96, 96)),  # Smaller size to speed up training
        transforms.Grayscale(num_output_channels=3),  # Convert to 3 channels
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    train_dataset = QuickDrawDataset(X_train, y_train, transform=transform)
    val_dataset = QuickDrawDataset(X_val, y_val, transform=transform)
    test_dataset = QuickDrawDataset(X_test, y_test, transform=transform)

    model = MobileNetRNNHybrid(num_classes=len(CLASS_NAMES), use_pretrained=True).to(device)
    print("Model initialized:")



    total_params = sum(p.numel() for p in model.parameters())
    print(f"Total parameters: {total_params:,}")

    print(model)

    print("\nPhase 1: Training only the RNN branch and classifier layers...")
    for param in model.mobilenet.parameters():
        param.requires_grad = False

    model, _, _, _, _ = train(
        model,
        train_dataset,
        val_dataset,
        device,
        batch_size=32,
        num_epochs=5,
        learning_rate=0.01
    )


    print("\nPhase 2: Fine-tuning ...")
    for param in model.mobilenet.parameters():
        param.requires_grad = True

    model, train_losses, train_accs, val_losses, val_accs = train(
        model,
        train_dataset,
        val_dataset,
        device,
        batch_size=32,
        num_epochs=5,
        learning_rate=0.001
    )

    # load best model
    best_model_path = f"{model.name}_best.pth"
    if os.path.exists(best_model_path):
        model.load_state_dict(torch.load(best_model_path))
        print(f"Loaded best model from {best_model_path}")

    test_loader = DataLoader(test_dataset, batch_size=32, shuffle=False, collate_fn=collate_fn)

    # evaluate on test set
    test_accuracy = evaluate(model, test_loader, device)
    print(f"\nTest Accuracy: {test_accuracy:.2f}%")

    #  confusion matrix
    cm = plot_confusion_matrix(model, test_loader, CLASS_NAMES, device)

    # Compare with baseline (MobileNetV2 only)
    print("\nComparing with baseline model (MobileNetV2 only)...")

    #image-only dataset
    class ImageOnlyDataset(Dataset):
        def __init__(self, stroke_data, labels, transform=None):
            self.stroke_data = stroke_data
            self.labels = labels
            self.transform = transform

        def __len__(self):
            return len(self.labels)

        def __getitem__(self, idx):
            stroke_seq = self.stroke_data[idx]
            label = self.labels[idx]
            img = strokes_to_image(stroke_seq)
            if self.transform:
                img = self.transform(img)
            return img, label

    # Create baseline datasets
    train_dataset_baseline = ImageOnlyDataset(X_train, y_train, transform=transform)
    val_dataset_baseline = ImageOnlyDataset(X_val, y_val, transform=transform)
    test_dataset_baseline = ImageOnlyDataset(X_test, y_test, transform=transform)

    # initialize baseline model
    baseline_model = models.mobilenet_v2(pretrained=True)
    baseline_model.name = "mobilenet_v2_baseline"
    baseline_model.classifier = nn.Linear(baseline_model.last_channel, len(CLASS_NAMES))
    baseline_model = baseline_model.to(device)


    baseline_best_path = f"/content/drive/MyDrive/APS360 Project/*Models and Code/{baseline_model.name}_best.pth"
    if os.path.exists(baseline_best_path):
        baseline_model.load_state_dict(torch.load(baseline_best_path, map_location=device))
        print(f"Loaded pre-trained baseline model from {baseline_best_path}")

    # Evaluate baseline on test set
    test_loader_baseline = DataLoader(test_dataset_baseline, batch_size=32, shuffle=False)
    baseline_model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in tqdm(test_loader_baseline, desc="Evaluating baseline"):
            images, labels = images.to(device), labels.to(device)
            outputs = baseline_model(images)
            _, predicted = outputs.max(1)
            total += labels.size(0)
            correct += predicted.eq(labels).sum().item()

    baseline_test_acc = 100. * correct / total
    print(f"Baseline Test Accuracy: {baseline_test_acc:.2f}%")
    print(f"Hybrid Model Test Accuracy: {test_accuracy:.2f}%")
    print(f"Improvement: {test_accuracy - baseline_test_acc:.2f}%")

    return {
        'hybrid_model': model,
        'baseline_model': baseline_model,
        'test_accuracy': test_accuracy,
        'baseline_test_accuracy': baseline_test_acc,
        'improvement': test_accuracy - baseline_test_acc
    }

if __name__ == "__main__":
    try:
        from google.colab import drive
        drive.mount('/content/drive')
        print("Google Drive mounted successfully")
    except:
        print("Not running in Google Colab or drive already mounted")

    # Run  model
    results = run_model()
```
